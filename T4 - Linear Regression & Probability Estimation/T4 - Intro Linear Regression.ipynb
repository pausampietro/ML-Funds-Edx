Trying to fit a Linear model using 1 or many features:

* **Eliminating the `b` coeficient by integrating it in `w`**

In order to have a more compact/workable model, we can convert the `b` parameter in another `w`, by adding a new feature which is identically 1 always, for every point. And then, multiplying it by a `b` which becomes a new `w_i` compoment of the model.

### Regularized Linear Regression (Ridge Regression)

If n is large enough, training loss (error) could be a good estimator of future performance of the model. Otherwise, we will probably be underestimating the performance over the test set

If `training error` is not a good estimator of `test error`, how we estimate the test error instead? 

1. **Adding a "penalty term" called regularization**
2. **k-fold cross-validation**

#### > Adding Regularization:

Adding a "penalty term" means adding a `lambda` to the model. The higher is the value of the paramater `lambda`, the less atention (weight) we are paying to the training set, and the lower are the `w` values. If we set a `lambda` too high, then it won't work because we are losing traceability of the impact of features (`w` almost zero).

Depending on the way the distance is computed, we will have many tipes of regularization. Lasso is tipical when the computed distance is L1 distance.

But it has one additional big added benefit, which is that it tends to produce sparse solutions w. That is to say it produces vectors w that tends to have a lot of zeros in it, just a few nonzero entries. **With Lasso `w` is sparse**, which have computing benefits: 
- Model is easier to understand, 
- Less computing-time spent
