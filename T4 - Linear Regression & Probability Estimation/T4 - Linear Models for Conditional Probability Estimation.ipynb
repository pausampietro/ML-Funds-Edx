## T4 - Linear Models for Conditional Probability Estimation

### Uncertainty in Prediction

Building a classifier, it could seem that having enought (infinite) training data, we can get as a result a perfect classifier. But, unfortunately, it is not like that. But why?

1. Because having all the data doesn't means having enough information to predict `y`. (For example, having all the medic information from patients doesn't allow us to predict exactly whether they will get a disease or not)
2. Limitations of the model class: if the "real boundary" is not linear, a linear approach will always have some error.


### Conditional probability estimation for binary labels

We define a linear boundary, which represents de 50% of probability of a point belonging to one class or another. Then, parallel lines from both sides represents more or less probability of belonging to class1/2. Obviusly probability of belonging to 1: `p1 = 1 - p2`

all we're gonna do is to compute this linear function. In the following example, 

$$ x1+x2-2 = 0$$

If it works out to zero, there's a 50, 50 chance.
If it's positive, then the chance is greater than half.
If it's negative, the chance is less than a half.
And the larger this value, the larger the probability that y equals one.

Now, to convert the distance from this parallel lines to probabilities that estimate the class, we will be using **the squasing function**. What this function does is converting a number between minus infinity and infinity (right side of equation of the lines) into a number between 0 and 1 (probability).
