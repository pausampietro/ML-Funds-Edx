## T1 - Distance Functions

3 families of Distance Functions:
1. Lp Norms
2. Metric Spaces
3. Non-metric Distance functions

##### > 1. Lp Distances:

Euclidean distance: **(SUM[i,n, (xi-zi)^2])^1/2**

Where p = 2. For a generic p, we have: **(SUM[i,n, (xi-zi)^p])^1/p**

- L1 distance: sum of the cartesian distances
- L2 distance: pitagorean distance
- Linfinity distance: max of cartesian distances

##### > 2. Metric Spaces:

In space W, Distance Function satisfies: W x W -> R1
* d(x,y) >= 0 (non negativity)
* d(x,y) = 0 only if x = y
* d(x,y) = d(y,x) (symmetry)
* d(x,z) <= d(x,y) + d(y,z) (triangle inequality)

In the following example, these 2 pieces of DNA are pretty similar, we can just go from one to another ADDING or REMOVING a letter in one of the variables. Then, the distance is defined as **d = edit distance**

**!!! Edit distance:** number of insertions, deletions, and substitutions needed to get from x to y.
