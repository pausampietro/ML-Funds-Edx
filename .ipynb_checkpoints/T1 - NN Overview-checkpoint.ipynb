## T1 - Nearest Neighbor Classification

(_Supervised Learning_)
Data Set = Training set + Test set
   * Training images: x1, x2, x3...
   * Labels: y1, y2, y3...

_Classify a new item: Find it's nearest neighbor among Xi set, and assign Yi accordingly_
    

##### > Computing Distance in N dimensions:

Euclidian Distance: ||x-z|| =sqrt(SUM[i,N, (xi - zi)^2])

##### > Accuracy of nearest neighbor on MNIST:

_training error_: Zero (not a good indicator) overly optimistic predictor for future performance

_test error_: fraction of test point incorrectly classified (ex: 3,1%)

##### > Improving NN Classificator: k-NN

1. Find de "k" nearest neighbors: 1,3,5...
2. Return the most common label amongst them

 BUT, if there is no test set, how we estimate the error of k-NN for a particular k, in order to choose the better value of k??

##### > Cross-Validation

##### > Improving NN Classificator: Other Distances Functions

If an image is moved to the right respect another, for example, the euclidean representation (vector) will be significant different, so Euclidean distance is not a good fucntion for images

Some options:
* Tangent Distance (same set, error: 1,1%)
* Shape Context (same set, error: 0,6%)

#### > Related Problem: Feature Selection

It is important to identify what features are involved in distance calculus, because wrong feature selection could dominate the clasificator ending in a bad model.

#### > Related Problem: Computing Time

Naive search takes time O(n) for training set of size n: SLOW!

Luckily there are Datra structures for speeding up NN search, like: 
* Locality sensitive hashing
* Ball Trees
* K-d trees
