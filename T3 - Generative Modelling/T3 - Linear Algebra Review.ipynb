## T3 - Linear Algebra Review

### Data as Vectors and Matrices

To represent data we could use vectors and matrices:

### Basic properties and operations:

* Transposed matrix:
$$ (A_{ij})^T = A_{ji} $$
$$ ((A)^T)^T = A $$

* Inverse matrix (for square matrices that are `invertible` = `not singular`)

$$ AA^{-1} = A^{-1}A = Id $$

* Adding and substracting: element by element
* Dot product of 2 vectors: 
$$ xy = x1y1 + x2y2 + ... $$

The dot product indicates the angle between the two vectors. If the dot product is zero: `x` and `y`are perpendicular, if the dot product (once normalized) is 1, it means they are paralel 

### Linear functions and dot products

Linear separator (boundary for classification in d dimensions): for `x = (x1,x2,...xd)` linear separators are of the form `wd = c`:
$$ w1x1 + w2x2 + ... wdxd = c $$

* **Basic dot product:**

A linear function from `R^4` to `R`: for example the dot product:

$$ f(x1,x2,x3,x4) = (3,0,-2,0)  (x1,x2,x3,x4) = 3x1-2x3 $$

* **Generalization for d dimensions:**

A linear function from  from `R^4` to `R^3`: performed with a matrix operation: 

$$ f(x1,x2,x3,x4) = (4x1 - x2, x3, -x1 + 6x4) $$
