### Multiclass discriminant analysis

When there are more than 2 classes, for example `k` classes, every class has its own quadratic function, depending on the probability of this class and the density funciton i d-dimensions.

When a new "point" appears, every class evaluates their quadratic function, and the points gets classified to the class with the highest result.
