## T2 - Probability Review

### Probability Space:

From rolling 2 Dice, for example, what are the probability of them adding up 10? we have:
- Sample Space: $$Omega = ((1,1),(1,2)...(2,1),(2,2),..) = (1,2,3,4,5,6) x (1,2,3,4,5,6) = (1,2,3,4,5,6)^2$$

- Probabilities of outcomes: _each outcome: 1/36_
- Event of interest: $$A <= Omega$$
$$ A = (z1,z2): z1+z2 = 10$$

In our example, events of interest: (4,6), (5,5), (6,4), so the probability of the fact is 3 · 1/36 = **1/12**


### Multiple Events: Conditional Probability

_Example:_ 10 coins, 9 are fair and 1 of them always comes up tails. You pick a coin at random. You toss it 4 times, and it comes up tails every time. <font color ="orange">What is the probability you picked the bad coin?</font>

* Sample space: (Coin #, toss 1, toss 2, toss 3, toss 4): (_Omega: 160 possible cases_)

$$ Omega = (1,2,...,10) x (H,T) x (H,T) x (H,t) x (H,T)$$
$$ Omega = (1,2,...,10) x (H,T)^4 $$


##### 1. Event A: Picked the bad coin
$$ A = (10, ?, ?, ?, ?) $$

##### 2. Event B: All coins are tails
$$ B = (?, T, T, T, T) $$

##### 3. Probability of A and B (both events occurs)

Probability of A **and** B, so probability of A **intersection** B:

$$ P(AintB) = P(bad.coin)·P(all.tails|bad.coin) $$
$$ P(AintB) = 1/10 · 1 = 1/10 $$


##### 4. Probability of A given B: 

$$ P(A|B) = P(AintB)/P(B) $$
$$ P(A|B) = 1/10 / P(B) $$

 Probability of B: All tails, it depends on wheter or not we have chosen the bad coin, so we sum both cases: 
 * P(all tails) = P(bad coin, all tails) + P(good coin, all tails) = 1/10 * 1 + 9/10 * 1/16 = 5/ 32

$$ P(A|B) = 1/10 / 5/32 = 0.64 $$

So there is a **64% probability** that we have picked up the bad coin
