### Probability Space:

From rolling 2 Dice, for example, what are the probability of them adding up 10? we have:
- Sample Space: $$Omega = ((1,1),(1,2)...(2,1),(2,2),..) = (1,2,3,4,5,6) x (1,2,3,4,5,6) = (1,2,3,4,5,6)^2$$

- Probabilities of outcomes: _each outcome: 1/36_
- Event of interest: $$A <= Omega$$
$$ A = (z1,z2): z1+z2 = 10$$

In our example, events of interest: (4,6), (5,5), (6,4), so the probability of the fact is 3 · 1/36 = **1/12**


### Multiple Events: Conditional Probability

_Example:_ 10 coins, 9 are fair and 1 of them always comes up tails. You pick a coin at random. You toss it 4 times, and it comes up tails every time. <font color ="orange">What is the probability you picked the bad coin?</font>

* Sample space: (Coin #, toss 1, toss 2, toss 3, toss 4): (_Omega: 160 possible cases_)

$$ Omega = (1,2,...,10) x (H,T) x (H,T) x (H,t) x (H,T)$$
$$ Omega = (1,2,...,10) x (H,T)^4 $$


##### 1. Event A: Picked the bad coin
$$ A = (10, ?, ?, ?, ?) $$

##### 2. Event B: All coins are tails
$$ B = (?, T, T, T, T) $$

##### 3. Probability of A and B (both events occurs)

Probability of A and B, so probability of A **intersection** B:

$$ P(AintB) = P(bad.coin)·P(all.tails|bad.coin) $$
$$ P(AintB) = 1/10 · 1 = 1/10 $$


##### 4. Probability of A given B: 

$$ P(A|B) = P(AintB)/P(B) $$
$$ P(A|B) = 1/10 / P(B) $$

 Probability of B: All tails, it depends on wheter or not we have chosen the bad coin, so we sum both cases: 
 * P(all tails) = P(bad coin, all tails) + P(good coin, all tails) = 1/10 * 1 + 9/10 * 1/16 = 5/ 32

$$ P(A|B) = 1/10 / 5/32 = 0.64 $$

So there is a **64% probability** that we have picked up the bad coin

### Random Variables

A random variable is defined on a probability space, it is a mapping from `Omega` to `R`. 
- Example: given the probability space of rolling two dices, lets take as a random variable `X` their sum.
- Example: rolling a dice, define `X=1`if the result is >=3, `X=0` otherwise.

#### Expected Value of a Random Value: `E(x)` -> `mu`

The expected value of a random variable `X` is its mean.
- Example: expected value of `X`: Roll a dice

$$ E(X) = 1·1/6+2·1/6+3·1/6+... = 3.5 $$
- Example: a biased coin has heads probability p. Expected value for `X`: 1 if heads, 0 if tails:

$$ E(x) = 0·(1-p) + 1·p = p $$

#### Expected value of distance to the mean: `E([x-mu]^2)` -> `var(x)`

The expected distance from the variable to the mean is called variance.


### Independent Random variables

2 random variables ara independent if the probability of both occuring: 

$$ P(X=x, Y=y) = P(X=x) · P(Y=y)$$

_Example:_ Flip a fair coin 10 times. Lets take `X`: # of heads, `Y`: last is a toss. These two variables doesn't seem to be independent, since the major number of heads seems to be influencing the probability of having a last toss. Let's check it:

$$ P(X=10, Y=T) = 0 $$
but
$$ P(X=10) = 1/2·1/2·1/2... = 1/2^{10} $$
$$ P(Y = toss) = 1/2 $$

So these variables are **not** independent, since the product of the two probabilities separately is not the probability of both at the same time.

We can check in this case, that the probability of every cell in the table is the product of each occuring separately, so there are **independent** random variables

### Correlation Coeficient

Correlation coeficient `r: [-1,1]` is a number that indicates how correlated two variables are.


**> Covariance:** Also indicates how related are the two variables, and it can be calculated with the following formula: 

$$ cov(X,Y) = E(X·Y) - E(X)·E(Y) $$

Which is in the range `[-std(X)·std(y), std(X)·std(Y)]`

**> Correlation:** is the normalized value of covariance, by dividing the covariance to normalize it

$$ corr(X,Y) = cov(X,Y) /  (std(X)·std(Y)) $$

Which is always in the range `[-1, 1]`
