## T2 - Generative Approach to Classification

### Generative models:

As an example, given a 1-dimension continuous variable `X`, we can define 3 Classes/labels `Y`


Then, we have:
1. the general probability of each Class (# items in class j /total items) 
$$pi(class_y)$$
2. The probability distribution of that class:
$$ P_y(x) $$

Then, to classify a new x: we have to calculate the **P(x,y):probability of finding x in label y**:   

$$ P(x,y) = pi(class_y)· P_y(x) $$

Then we have to pick up the class with the largest `P(x,y)`

### Fitting a Generative Model in 1 dimension

_Example:_ Assign the winery of a wine that have lost the label. _Training set: 130 bottles_

- Winery 1: 43 bottles
- Winery 2: 51 bottles
- Winery 3: 36 bottles

With these numbers, we calculate `pi_1 = 0.33`, `pi_2 = 0.39` , `pi_3 = 0.28`. 

For each bottle we have 13 features, but we will start classifying with 1 feature: `Alcohol (%)`.
To Find the largest `pi(class_y)· P_y(x)`, the steps to follow are:

1. With the alcohol measures, create an _histogram_
2. Calculate `u` as a mean and `sigma^2` for a normal distribution (or other probabilistic distribution) 
3. Once having the 3 normal distribution for each feature (13 distributions x 3 wineries), we calculate the probability of having `alcohol_%_wine_x` in each, and ponderate with `pi_y`

Result: Using one feature ('Alcohol'), error rate is `29%`
